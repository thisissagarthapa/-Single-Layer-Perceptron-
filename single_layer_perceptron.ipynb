# Single Layer Perceptron

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import LabelEncoder
```

```python
dataset=pd.read_csv("salary_classification_data.csv")
```

```python
dataset.head()
```

```python
dataset.drop(columns=["Age"],inplace=True)
```

```python
le=LabelEncoder()
le.fit_transform(dataset["Salary_Category"])
dataset["Salary_Category"]=le.fit_transform(dataset["Salary_Category"])
```

```python
dataset.isnull().sum()
```

```python
dataset.head(4)
```

```python
sns.scatterplot(x="Years_of_Experience",y="Salary",data=dataset,hue="Salary_Category")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Salary_Category"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr=Perceptron()
```

```python
pr.fit(x_train,y_train)
```

```python
pr.score(x_train,y_train)*100,pr.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
plt.show()
```
